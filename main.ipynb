```python
# Analyze model input
data_dir = './dataset/dataset_prep2.csv'
dataset = load_dataset('csv', data_files={'train': data_dir})
dataset = dataset.map(
            lambda example: tokenizer(example['text'], max_length=512, padding='max_length', truncation=True, return_tensors="pt"),
            batched=True
            )
dataset = dataset.shuffle()

for i in range(1):
    dd = dataset['train'][i]
    input_ids = dd['input_ids']
    label = dd['label']
    attention_mask = dd['attention_mask']
    text = tokenizer.decode(input_ids)

    print('input_ids: ',input_ids)
    print('att_mask: ', attention_mask)
    print('inp * mask: ', [a*b for a,b in zip(input_ids,attention_mask)])
    print('text: ', text)
    print('len(input_ids): ', len(input_ids))
    print('label: ', label)
    print('\n\n')
```
